```python
import numpy as np
import plotly.graph_objects as go

# Your array
arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# Convert to 3D coordinates
x, y = np.meshgrid(np.arange(arr.shape[1]), np.arange(arr.shape[0]))
x = x.flatten()
y = y.flatten()
z = arr.flatten()

# Frames for animation (each frame adds one point)
frames = [
    go.Frame(data=[
        go.Scatter3d(
            x=x[:i+1],
            y=y[:i+1],
            z=z[:i+1],
            mode="markers",
            marker=dict(size=8, color=z[:i+1], colorscale="Viridis")
        )
    ])
    for i in range(len(z))
]

# Base figure (first frame: only first point)
fig = go.Figure(
    data=[
        go.Scatter3d(
            x=[x[0]],
            y=[y[0]],
            z=[z[0]],
            mode="markers",
            marker=dict(size=8, color=[z[0]], colorscale="Viridis")
        )
    ],
    layout=go.Layout(
        title="Animated 3D Points from 2D Array",
        scene=dict(
            xaxis_title="X (Columns)",
            yaxis_title="Y (Rows)",
            zaxis_title="Z (Values)"
        ),
        updatemenus=[dict(
            type="buttons",
            showactive=False,
            buttons=[dict(
                label="Play",
                method="animate",
                args=[None, {"frame": {"duration": 500, "redraw": True},
                             "fromcurrent": True}]
            )]
        )]
    ),
    frames=frames
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

rows, cols = arr.shape
x = []
y = []
z = []

for i in range(rows):
    for j in range(cols):
        x.append(i)
        y.append(j)
        z.append(arr[i, j])

fig = go.Figure(data=[go.Scatter3d(
    x=x, y=y, z=z,
    mode='markers',
    marker=dict(size=8)
)])

fig.show()
```
